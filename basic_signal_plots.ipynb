# Basic Signal Plotting Using Python

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## Ramp Signal

```python
x = np.arange(0, 11, 1)
y = x
plt.plot(x, y)
plt.title("Plot of Ramp signal")
plt.xlabel("Values of x")
plt.ylabel("Values of y(x)")
plt.show()
```

## Signum Function

```python
x = np.arange(-11, 11, 0.001)
y = np.sign(x)
plt.plot(x, y)
plt.title("Plot of Sign(x)")
plt.xlabel("Values of x")
plt.ylabel("Values of sign(x)")
plt.show()
```

## log(x)

```python
x = np.arange(1, 11, 0.001)
y = np.log(x)
plt.plot(x, y)
plt.title("Plot of log(x)")
plt.xlabel("Values of x")
plt.ylabel("Values of log(x)")
plt.show()
```

## Sinusoidal Signal

```python
x = np.arange(-11, 11, 0.001)
y = np.sin(x)
plt.plot(x, y)
plt.title("Plot of sin(x)")
plt.xlabel("Values of x")
plt.ylabel("Values of sin(x)")
plt.show()
```

## Sinc Function

```python
x = np.arange(-11, 11, 0.001)
y = np.sinc(x)
plt.plot(x, y)
plt.title("Plot of sinc(x)")
plt.xlabel("Values of x")
plt.ylabel("Values of sinc(x)")
plt.show()
```

## Gaussian Pulse

```python
x = np.arange(-11, 11, 0.001)
y = np.exp(-x * x)
plt.plot(x, y)
plt.title("Plot of e^(-x^2)")
plt.xlabel("Values of x")
plt.ylabel("Values of e^(-x^2)")
plt.show()
```
